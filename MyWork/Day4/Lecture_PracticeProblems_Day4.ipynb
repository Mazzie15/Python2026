### Creating Numerical Lists for Loops

Many times you want a list of numbers, like all the numbers between 1 and 100. Typing this out would be a WASTE OF TIME! Luckily Python has our back...

```python
# Notice that range(#) starts at zero and goes to one less than the number
my_numbers = list(range(100))
print(my_numbers)
```

```python
# I want to start at 1 and go to 100!
my_numbers = list(range(1,101))
print(my_numbers)
```

```python
# This is really useful in lists
best_cat_names = ['Bongo','Tater Tot','Sofia','Bella','Muffin'] 
best_cat_colors = ['Black','Orange','Gray','Black','Gray']

# Here i will take number values that match the indexes of our list
# This means we can combine lists in a message.
for i in range(len(best_cat_names)):
    print(f"The number {i+1} cat, {best_cat_names[i]}, is a {best_cat_colors[i]} cat")
```

### Slicing Lists

Sometimes you don't want to use the whole list, just a small part of it. This is called slicing the lists.

```python
best_cat_names = ['Bongo','Tater Tot','Sofia','Bella','Muffin'] 

# Slicing goes from the first number to one less than the second number
print(f'The first three names in the list are: {best_cat_names[0:3]}')

print(f'The three middle names in the list are: {best_cat_names[1:4]}')

# You can use shortcuts if you are going from the beginning or to the end
print(f'The first three names in the list are: {best_cat_names[:3]}')
print(f'The last three names in the list are: {best_cat_names[-3:]}')
```

### Copying Lists - USER BEWARE!!!

```python
# Here are some of my favorite foods
my_favorite_foods = ['pizza','cookies','avocados','potatoes']
# My husband likes most of the same foods
ross_favorite_foods = my_favorite_foods
print(ross_favorite_foods)
```

```python
# But he doesn't like avocados
ross_favorite_foods.remove('avocados')
print(ross_favorite_foods)
```

```python
# Lets look at my favorite foods again
print(my_favorite_foods)
```

#### HEY! What happened to my avocados!!!

Beacuse I wrote:

    ross_favorite_foods = my_favorite_foods

Python has BOTH variables point at the same data. So when I canged Ross' favorite food it changed the ONE piece of data in memory, also changing my foods.

### Instead make a copy using slices

```python
# Here are some of my favorite foods
my_favorite_foods = ['pizza','cookies','avocados','potatoes']
# My husband likes most of the same foods
# but this time take a slice -- [:] takes the full list

ross_favorite_foods = my_favorite_foods[:]



# Now remove avocados
ross_favorite_foods.remove('avocados')

# And print both lists
print(ross_favorite_foods)
print(my_favorite_foods)
```

### Tuples

Another object to store data... just more strict


```python
# tuples are surrounded by parenthesis or round brackets
my_tuple = ('A','B','C',1,2,3)
# lists are surrounded by square brackets
my_list = ['A','B','C',1,2,3]

print(my_tuple)
print(my_list)
```

```python
# Access Elements - same
print(my_tuple[0])
print(my_list[0])
```

```python
# Access slices - same
print(my_tuple[0:3])
print(my_list[0:3])
```

```python
# For loops - same
for item in my_tuple:
    print(item)
```

#### BUT --- here's where they differ

```python
my_tuple[0]='D'
```

```python
my_list[0]='D'
print(my_list)
```

```python
print(my_tuple)
print(my_list)
```

Tuples should be used for data that should not be overwriten in a pice by piece way. Often these are used for dimensions, coordinates, locations (lattitued, longitude), etc. If you are writing code and you have some data that should not be changed one piece at a time then use a tuple.

#### How do we change the data in a tuple?

```python
# Overwrite the whole thing
my_tuple = ('D', 'B', 'C', 1, 2, 3)
print(my_tuple)
```
